# Advanced Indexing


```python
import numpy as np
```

```python
arr = np.array([10, 20, 30, 40, 50])
# condition
# use it to filter
print(arr>25)         # print true or false at each index
print(arr[arr>25])    # print value at index if true
```

```python
# multiple conditions
# AND -> &
# OR -> |
# NOT -> ~

print((arr > 10) & (arr < 40))
print(arr[(arr > 10) & (arr < 40)])
```

```python
# FANCY INDEXING
arr = np.array([10,20,30,40,])
print(arr[[0,2]])       # can pick random position directly
                        # print the values at index 0 and 2
print(np.random.choice(arr,3)) # randomly pick 3 values from array dont use .randint
```

```python
# Boolean Masking
marks = np.array([45,67,30,90])
# student who passed (> 50)
print(marks[marks > 50])
```

```python
# PRACTICE
dict = {
    "tarun": 65,
    "arun": 67,
    "varun": 90
}
for key, val in dict.items():
  if(val > 65):
    print(key)
```

```python
# AXIS CONCEPT

# Array Type	Valid Axis
# 1D	only axis=0
# 2D	axis=0, axis=1
# 3D	axis=0,1,2

arr = np.array([[1,2,3],
                  [4,5,6]])

# arr = np.array([1,2,3,4])
# “axis = 0 means column-wise, so output should be (1,2,3,4)”
# 👉 This logic works for 2D arrays, but not for 1D arrays
# "It Combine all elements along the only available axis”


print(np.sum(arr, axis=0))        # sum the elements column wise
np.sum(arr, axis=1)               # sum the elements row wise
```

# Advance NumPy level


```python
# Vector and matrices basics
import numpy as np

# vector
a = np.array([1,2,3])

# matrix
A = np.array([[1,2],
             [4,5]])
```

```python
# Dot Product
# Used in:
# Machine Learning
# Neural Networks
# Similarity calculations

a = np.array([1,2,3])
b = np.array([4,5,6])

print(np.dot(a,b)) # 1*4 + 2*5 + 3*6 = 32
```

```python
# Matrix Multiplication
A = np.array([[1,2],
              [3,4]])
B = np.array([[5,6],
              [7,8]])

print(np.dot(A,B))
# OR
print(A @ B)
```

```python
# Transpose
A = np.array([[1,2,3],
              [4,5,6]])
print(A.T)
```

```python
# Determinant
A = np.array([[1,2],[3,4]])
result = np.linalg.det(A)       # linalg-> linear algebra module inside numpy
                                # to perform mathema. operat. related to matrices and vectors

# Round value -> change the value
print(round(result, 7))     # Python removes unnecessary trailing zeros that why output is -2.0

# format display -> changes how the number looks
print(f"{result:.7f}")      # print upto 7 places after point
```

```python
# Inverse the matrix
A = np.array([[1,2],
              [3,4]])
inv_A = np.linalg.inv(A)
print(inv_A)

print(np.round(np.dot(A ,inv_A),2))
```

```python
# Eigen vectors and Values
# Used in:
# PCA (Dimensionality reduction)
# ML models
# Physics simulations

A = np.array([[2,0],
              [0,3]])
values, vectors = np.linalg.eig(A)
print(values)
print(vectors)
```

```python
# Linear Equation
# 2x + y = 5
# x + y = 3

A = np.array([[2,1],
              [1,1]])
B = np.array([5,3])

result = np.linalg.solve(A,B)   # solve the linear equation
print(result)
```

```python
# Norm (Vector length or magnitude of vector)
# root(x^2 + y^2 + .....)

# Used in:
# Distance calculation
# ML optimization

# DISTANCE BETWEEN POINTS
a = np.array([3,4])
b = np.array([5,6])

A = a - b
print(A)

distance = np.linalg.norm(A)
print(round(distance, 2))
print(f"{distance:.4f}")
```

```python
import numpy as np
A = np.array([[2,3],
             [1,4]])

B = np.array([[5,6],
              [7,8]])

print(A * B)  # Not A Matrix Multiplication
print(np.dot(A, B))
# OR
print(A@B)

print(f"{np.linalg.det(A):.2f}")

inv_a = np.linalg.inv(A)
print(inv_a)
print(np.round(np.dot(A, inv_a),2))
np.allclose(np.dot(A, inv_a), np.eye(2))

print(B.T)

x = np.array([[2,3],
              [1,2]])
y = np.array([8,5])

result = np.linalg.solve(x,y)
print(result)
```

```python
# crop image
# Similar patterns
# Code	Meaning
# img[:, 1:]	remove first column
# img[:, :-1]	remove last column
# img[1:, :]	remove first row
# img[:-1, :]	remove last row

cropped = img[:, 1:] # take all rows but remove the first column
print(cropped)
```

```python
# Flip image
print(img)
flipped = np.flip(img, axis = 1)      # columns (horizontal direction)
print(flipped)
```

```python
# It specifies the data type of the array elements
# 👉 int8 means: 8-bit signed integer
# What is “8-bit”?
# 👉 8 bits = 1 byte
# 👉 It can store 256 values
# Since it’s signed:
# Range: -128 to 127

# Memory Efficient
# int8  → 1 byte per element
# int32 → 4 bytes per element
# int64 → 8 bytes per element

arr = np.array([1,2,3], dtype = np.int8)
# arr = np.array([1,2,3], dtype = np.int64)
print(arr.nbytes)
```

```python
# Find duplicate elements
import numpy as np

arr = np.array([1,2,3,2,4,5,1])
unique , count = np.unique(arr, return_counts = True) # This does two things:
                                                      # Finds unique values
                                                      # Counts how many times each value appears
duplicates = unique[count > 1]
print(duplicates)
```

```python
# Normalize Data
arr = np.array([10,20,30,40,50])
norm = (arr - np.min(arr)) / (np.max(arr) - np.min(arr))
print(norm)
```

```python
# Replace value based on condition
arr = np.array([10,25,30,25,5])
result = np.where(arr > 10, 0, arr)
print(result)
```

```python
# Row-wise sum
arr = np.array([[1,2,3],
                [4,5,6]])
result = np.sum(arr ,axis = 0)
result1 = np.sum(arr, axis=1)
print(result)
print(result1)
```

```python
# Find Top K elements
arr = np.array([10,40,30,20,50])
k = 3
top_k = np.sort(arr)[-k:]
print(top_k)
```

```python
# Matrix multilication
a = np.array([[1,2],
              [3,4]])
b = np.array([[5,6],
              [7,8]])
print(np.dot(a,b))
```

```python
# Flatten Array
arr = np.array([[1,2,3,45],
               [4,5,6,7]])
print(arr.flatten())
```

```python
# count even
arr = np.array([1,2,3,4,5,6])
count = 0
for val in arr:
  if val % 2 == 0 :
    count += 1

print(count)

# OR

new_count = np.sum(arr%2 == 0)
print(new_count)
```
