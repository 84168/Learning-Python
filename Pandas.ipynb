Intermediate Pandas

```python
# Conditional Filtering 
import pandas as pd
data = {
    "name": ["tarun","arun","varun"],
    "marks": [20,50,80]
}
df = pd.DataFrame(data, index=[10,20,30])
print(df)

df[(df["marks"] > 50) & (df["marks"]<90)]
```

```python
# Apply Function
# The .apply() method is a Pandas tool used to "apply" 
# a function along an axis of a DataFrame or a Series (a column).

# A lambda function is a small, anonymous function. {lambda arguments: expression}
# Unlike a standard function defined with def, a 
# lambda has no name and is contained in a single line of code.

df["marks"].apply(lambda x: x+5)
```

```python
df["status"] = df["marks"].apply(lambda x : 'Pass' if x >20 else 'Fail')
print(df)
```

```python
def msg(marks):
    if(marks <= 20):
        return "can do better"
    elif(marks > 20 and marks <= 50):
        return "can do good"
    else:
        return "excellent"
# In standard Python, you call a function immediately: msg(20). 
# But in Pandas, you aren't calling the function; you are passing the function.
df["msg"] = df["marks"].apply(msg)
print(df)
```

```python
# Sorting Multiple Columns

# by = [primary sort , secondary sort]
# if primary tie break then we sort on the basis of secondary sort
# ascending = [false, true] primary sort in descending and secondary in ascending order

df.sort_values(by= ["marks","name"], ascending=[False, True])
```

```python
# Rename columns
# df.rename() used to change labels allows to change hust one specific name 
# columns = {old_col_name, new_col_name} can change multiple col name simultaneously like:
                # columns = {old1:new1, old2:new2, .....}
# inplace , if Flase(default) it only shows the preview of result original df remain unchanged
#           if True (manual) it change the df originally

df.rename(columns={"marks": "score"}, inplace = True)
print(df)
```

```python
df.loc[40] = ["kiran", 75, 'pass', "can do better"]
print(df)
```

```python
# Drop column / rows
df.drop("score", axis = 1) # column after delete it shows the preview cause no inplace is true here
df.drop(40, axis = 0) # row same here it is the preview
```

```python
# Merge (sql join)
df1 = pd.DataFrame({
    "id":[1,2],
    "name":["A","B"]
})

df2 = pd.DataFrame({
    "id":[1,2],
    "marks":[80, 90]
})

pd.merge(df1,df2, on="id")                  # Inner join
pd.merge(df1,df2, on="id", how = "left")    # left join
pd.merge(df1, df2, on="id", how="right")    # right join
pd.merge(df1, df2, on="id", how="outer")    # outer join
```

```python
# df.pivot_table() , is summary tool of pandas shonks long data into meaning full report
df.loc[40] = ["tarun", 30, 'fail', "can do better"]
df.pivot_table(values="score", index="name", aggfunc="mean")
print(df)
```

```python
# Handling duplicated
df.loc[50] = ["arun", 50, 'Pass',"can do good"]
df.duplicated()

# Returns True/False for every row based ONLY on the "name" column
duplicates = df.duplicated(subset=["name"])
print(duplicates)

# To actually see the duplicate rows:
print(df[df.duplicated(subset=["name"])])
```

```python
# String operation
df["name"].str.upper()
df["name"].str.contains("A") # Case sensitive, return true or false 

# print(df)
```

```python
import pandas as pd

data = {
    "name": ["A", "B", "C", "A"],
    "marks": [85, 40, 90, 85],
    "dept": ["IT", "HR", "IT", "IT"]
}

df = pd.DataFrame(data)

df.drop_duplicates()
df["result"] = df["marks"].apply(lambda x: 'Pass' if x > 40 else 'Fail')
print(df)

df.groupby("dept")["marks"].mean()

df.sort_values(by= ["marks"], ascending = False)
print(df["name"].str.lower())
```
